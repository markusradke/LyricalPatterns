```python
import pickle 
import matplotlib.pyplot as plt
```

```python
def report_results_for_model(folder):
    metadata = pickle.load(open(f"../models/{folder}/stm_metadata_2-30.pkl", "rb"))
    chosen = metadata["K"]
    plt.plot(metadata["search_results"]["K"], metadata["search_results"]["heldout"])
    plt.axvline(chosen, color = "red")
    print(f"chosen number of topics: {chosen}")
    return metadata["search_results"]

def get_first_local_max_index(search):
    heldout, K = search["heldout"], search["K"]
    first_local_max = next(i for i in range(1, len(heldout)-1) if heldout[i] > heldout[i-1] and heldout[i] > heldout[i+1])
    topics = K[first_local_max]
    print(f"first local likelihood max of topics: {topics}")
    return topics
```

```python
search = report_results_for_model("stm_topics_dc")
first = get_first_local_max_index(search)
plt.axvline(first, color="green")
plt.axvline(16, color="orange")
```

```python
search = report_results_for_model("stm_styles_dc")
first = get_first_local_max_index(search)
plt.axvline(first, color="green")
plt.axvline(16, color="orange")
```
